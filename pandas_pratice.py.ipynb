# Cleaning data

```python
#removing the empty cells:this code change only in new file but original file is asits


import pandas as pd
df =pd.read_csv("data.csv")

new_df=df.dropna()
print(new_df.to_string())
```

```python
#this changes also orignal dataframe
import pandas as pd 
df=pd.read_csv("data.csv")
df.dropna(inplace=True)

print(df.to_string()) 
```

```python
#filling the empty cells by 130:
import pandas as pd 
df=pd.read_csv("data.csv")
df.fillna(130,inplace=True)
print(df.to_string())
```

```python
#replacing at specific positon

import pandas as pd

df=pd.read_csv("data.csv")

df["Calories"].fillna(130,inplace=True)

print(df.to_string)
```

```python
#replacing value by median

import pandas as pd
df=pd.read_csv("data.csv")

x=df["Calories"].median()

df["Calories"].fillna(x,inplace=True)

print(df.to_string())
```

```python
#mean

    
import pandas as pd
df=pd.read_csv("data.csv")

x=df["Calories"].mean()

df["Calories"].fillna(x,inplace=True)

print(df.to_string())
```

```python
#mode()

import pandas as pd
df=pd.read_csv("data.csv")

x=df["Calories"].mode()

df["Calories"].fillna(x,inplace=True)

print(df.to_string())
```

# Cleaning data for wrong format:

```python
#datetime
import pandas as pd

df=pd.read_csv("data.csv")
df["Date"]=pd.to_datetime(df["Date"])

#print(df.to_string) this code will be not execute because of  data column not in our orignal csv file
```

```python
#at  index 7 and column of "Duration" is change as 45 fix value
import pandas as pd

df=pd.read_csv("data.csv")

df.loc[7,"Duration"]=45

print(df.to_string())
```

```python
#in below code we change the all duration value which are above 120 we make them 120
import pandas as pd

df=pd.read_csv("data.csv")
for x in df.index:
    
    if df.loc[x,"Duration"] >120:
        
        df.loc[x,"Duration"]=120
    

print(df.to_string())
```

```python
#removing the column which is above 120
import pandas as pd
df=pd.read_csv("data.csv")
for x in df.index:
    
    if df.loc[x,"Duration"] >120:
        
        df.drop(x,inplace=True)
    

print(df.to_string())
```
